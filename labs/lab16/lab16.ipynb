# problem 1

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def polynomial(x):
    return (np.linalg.solve(np.vander(x), f(x)))
```

```python
x = np.linspace(1, 15, 100)
y = f(x)
```

```python
deg_1 = polynomial(np.array([1, 15]))
deg_2 = polynomial(np.array([1, 8, 15]))
deg_3 = polynomial(np.array([1, 4, 10, 15]))

plt.figure(figsize=(15, 10))
plt.plot(x, y, lw=3, label="solution")
plt.plot(x, np.polyval(deg_1, x), label="deg_1")
plt.plot(x, np.polyval(deg_2, x), label="deg_2")
plt.plot(x, np.polyval(deg_3, x), label="deg_3")
plt.grid(which='both')
plt.legend()
plt.show()
```

# problem 2

```python
def strange_function(x, y):
    print(np.count_nonzero(x*y))
    print(np.count_nonzero(abs(x) + abs(y)))
```

```python
x = np.random.randint(-5, 5, size = 10)
y = np.random.randint(-5, 5, size = 10)
print(x)
print(y)
strange_function(x, y)
```

# problem 3

```python
from random import randint
```

```python
x = np.random.randint(-5, 5, size = 10)
n = randint(2, 5)
print(x,", N = ", n)
```

```python
def repeating_function(x, n):
    mat = x + np.zeros((n, 1)).astype(np.int) 
    return mat.T.reshape(len(x)*n)
```

```python
repeating_function(x, n)
```
